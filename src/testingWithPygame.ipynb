```python
import random
import math
from PyQt6.QtCore import QSize
import pygame as pg
import numpy as np
from Singleton import Singleton as S
from Point import Point
from Cope import *

# constants
WINSIZE = [640, 480]
WINCENTER = [320, 240]
NUMSTARS = 150

white = 255, 240, 200
black = 20, 20, 40

home = np.array([
    [16, 0, WINSIZE[0] / 2],
    [0, 16, WINSIZE[1] / 2],
    [0, 0, 1],
    ])
transformation = home.copy()
relativeTransformation = transformation.copy()
```

```python
xarr = []
yarr = []

def genDots(vertical=False):
    if vertical:
        if True:
            #                                   x and y scale factor
            xAmt = round((WINSIZE[0]  / 2) / int(transformation[0, 0]))
            yAmt = round((WINSIZE[1] / 2) /  int(transformation[1, 1]))
            for x in range(-xAmt, xAmt):
                for y in range(-yAmt, yAmt):
                    xarr.append(x)
                    yarr.append(y)
            dots = np.array([xarr, yarr, [1] * len(xarr)])
        else:
            area = Point(WINSIZE) / 2
            x, y = np.mgrid[
                -area.x : area.x : int(transformation[0, 0]),
                -area.y : area.y : int(transformation[1, 1])
            ]
            # They should have the same shape
            rows = len(x[0])
            cols = len(y)
            for r in range(rows):
                for c in range(cols):
                    xarr.append(x[c, r])
                    yarr.append(y[c, r])
            dots = np.array([xarr, yarr, [1] * len(xarr)])
    else:
        pass
    return dots
    
genDots()
```

```python
def drawDots(surface, dots, color=white):
    for pos in dots.T:
        pos = (int(pos[0]), int(pos[1]))
        surface.set_at(pos, color)
```
